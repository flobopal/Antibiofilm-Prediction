```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Antibiofilm MF+MD.csv", index_col=0)
data.head()
```

```python
data.loc[data.target_organism.str.startswith("Salmonella"), "target_organism"] = "Salmonella enterica"
```

```python
data.describe()
```

```python
data.pIC50.plot(kind="hist", bins=25, colormap = "gist_gray", edgecolor="white")
plt.xlabel("pIC50")
```

```python

plt.figure(figsize=(6,2))
ax = data["target_organism"].value_counts().plot(kind="barh", colormap="gist_gray", width=0.8)
plt.ylabel("Target Organism")
plt.xlabel("Data size")
ax.bar_label(ax.containers[0], label_type="center", color="w", )
ax.set_xlim(right=450)
```
