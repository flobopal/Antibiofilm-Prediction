```python
import pandas as pd
import os
```

```python
predictions = None

for file in sorted(os.listdir()):
    if os.path.isdir(file) and file != "old":
       df = pd.read_csv(os.path.join(file, "predictions.csv"), index_col=0)
       model_index = file.split()[0]
       if predictions is None:
           predictions = df
           predictions.rename(columns={"y_pred": f"model {model_index}"}, inplace=True)
           continue
       predictions[f"model {model_index}"] = df.y_pred

predictions.to_csv("predictions.csv")
predictions.head()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

metrics = pd.DataFrame(index=["R2", "MAE", "RMSE"])
for column in predictions.columns[-6:]:
    metrics[column] = [
        r2_score(predictions.pIC50, predictions[column]),
        mean_absolute_error(predictions.pIC50, predictions[column]),
        mean_squared_error(predictions.pIC50, predictions[column])**0.5
    ]

metrics.T.round(3)
```
